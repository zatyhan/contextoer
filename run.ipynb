```python
import gensim.downloader as api
import nltk 
from nltk.corpus import words
from nltk.stem import WordNetLemmatizer
```

```python
from splinter import Browser
from selenium.webdriver.common.keys import Keys
import random
```

```python
api.info()['models'].keys()
```

```python
model = api.load("word2vec-google-news-300")
```

```python
nltk.download('all')
```

```python
vocabs= list(model.key_to_index.keys())

lemmatizer= WordNetLemmatizer()

english_words= set(words.words())

filtered_words = [] # list of words that are in the model and is valid in the English language
for word in vocabs:
    lemma = lemmatizer.lemmatize(word, pos='n')

    # will need to check whether same words of mixed capitalization counts toward the same word 
    if lemma == word and word.lower() in english_words:
        filtered_words.append(word.lower())
```

```python
print(len(filtered_words))
print(len(set(map(str.lower, filtered_words))))
sorted(map(str.lower, filtered_words))
```

```python
for word in ['sheep', 'cow', 'donkey', 'alligator']:
    if word in model:
        print(f"{word} is in the model")
```

```python
# store the embeddings of each word

import numpy as np
embeddings = np.stack([model[word] for word in filtered_words if word in model])
len(embeddings)
```
